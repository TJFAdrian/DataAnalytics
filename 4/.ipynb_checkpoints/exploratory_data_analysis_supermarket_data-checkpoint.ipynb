# Exploratory data analysis (EDA) of supermarkets data

## Libraries and settings

```python
# Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# seaborn graphics settings
sns.set(color_codes=True)

# Ignore warnings
import warnings
warnings.filterwarnings("ignore")
```

## Univariate non-graphical exploratory data analysis (EDA)

### Importing the enriched apartment data

```python
# Read and select variables
df = pd.read_csv("supermarkets_data_enriched.csv")[['id',
                                                   'brand',
                                                   'lat',
                                                   'lon',
                                                   'pop',
                                                   'pop_dens',
                                                   'frg_pct',
                                                   'emp']]
```

```python
# Count amount of stores per brand
df['brand'].value_counts()
```

```python
cols = set(df.columns) - {'id'}
df1 = df[list(cols)]
df1.describe()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Plot Histogram
fig = plt.figure( figsize=(10,8))
plt.xticks(fontsize=14, rotation=0)
plt.yticks(fontsize=14, rotation=0)
n, bins, patches = plt.hist(x=df['pop_dens'], 
                            bins=40, 
                            color='#42AD12',
                            alpha=0.5, 
                            rwidth=0.7
                   )
plt.grid(True)
plt.ticklabel_format(style='plain')
plt.grid(axis='y', alpha=0.75)

# Set labels
plt.xlabel('pop_dens', fontsize=10, labelpad=10)
plt.ylabel('Frequency', fontsize=10, labelpad=10)
plt.title('Histogram of pop_dens', fontsize=12, pad=10)

# Set fontsize of tick labels
plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)

plt.show()
```

```python
sns.set(style="ticks", font_scale=0.8)
g = sns.PairGrid(df[['lat', 
                     'lon', 
                     'pop', 
                     'pop_dens',
                     'emp',
                     'frg_pct']], 
                 height=1.2, 
                 aspect=1)
g.map_upper(sns.scatterplot,color='darkgreen', s=10)
g.map_lower(sns.scatterplot, color='darkgreen', s=10)
g.map_diag(plt.hist, color='brown')
```

```python
df2 = pd.read_csv("supermarkets_data_enriched.csv")[['id',
                                                    'brand',
                                                    'lat',
                                                    'lon']]

# Subset
df_sub = df2.loc[df2['brand'].isin(['Coop', 'Migros', 'Denner', 'Volg', 'Landi'])]
df_sub

# Colors
colors = {'Coop':'bisque','Migros':'orange','Denner':'lightcoral','Volg':'lightsteelblue','Landi':'palegreen'}

# Plot
plt.scatter(df_sub['lon'],
           df_sub['lat'],
           c=df_sub['brand'].map(colors))
```
